```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn  as sns 
from seaborn import load_dataset
import numpy as np

# Load the iris dataset
iris =sns.load_dataset('iris')

# List down the features and their types
print("Features and their types:")
iris.dtypes
```

```python
# Create histograms for each feature
print("\nHistograms for each feature:")
iris.hist(figsize=(12, 8))
plt.show()
```

```python
# Create boxplots for each feature
print("\nBoxplots for each feature:")
iris.boxplot(figsize=(12, 8))
plt.show()
```

```python
#for SepalLengthCm
Q1 = np.percentile(iris['sepal_length'], 25)
Q3 = np.percentile(iris['sepal_length'], 75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
outliers = (iris['sepal_length'] < lower_bound) | (iris['sepal_length'] > upper_bound)
print("for sepal_length")
print("outlier:",outliers)
A=outliers.sum()
if A==0:
    print("no outlier")
else:
    print("outlier is present")
```
